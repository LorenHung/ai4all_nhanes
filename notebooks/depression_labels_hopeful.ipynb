## Import data

```python
all_data_df = pd.read_csv('all_data_df.csv')
all_data_df
```

```python
mental_health_df = all_data_df.loc[:, 'DPQ010':'DPQ100']
mental_health_df
```

## Categorize values
### Make a function

```python
def add(row):
    sum = 0
    for i in range(0, 10):
        sum += float(row[i])
    return sum

def categ(val):
    if val >= 0 and val <= 4:
        return "minimal"
    elif val >= 5 and val <= 9:
        return "mild"
    elif val >= 10 and val <= 14:
        return "moderate"
    elif val >= 15 and val <= 19:
        return "moderately severe"
    else:
        return "severe"
```

```python
mental_health_df['labels_raw'] = mental_health_df.apply(add, axis=1)

# tells if row was originally full of NaN
cols = mental_health_df.columns
mental_health_df['is_na'] = mental_health_df[cols].isnull().apply(lambda x: all(x), axis=1) 

mental_health_df
```

```python
mental_health_df = mental_health_df.fillna(0)
mental_health_df
```

```python
mental_health_df['original_labels'] = mental_health_df['labels_raw'].apply(lambda x: categ(x))
mental_health_df
```

```python
# replace 'minimal' with 'missing'
def rep(row):
    if row['is_na'] == True:
        return 'missing'
    else:
        return row['original_labels']

mental_health_df['labels'] = mental_health_df.apply(rep, axis=1)
mental_health_df
```

```python
mental_health_df.to_csv('mental_health_df.csv', index=False)
```
